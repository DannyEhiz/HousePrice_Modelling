<h3 style = "color:#FF2171"><b>Data Exploration and Cleaning</b></h3><hr>

```python
# Drop Column function 
def drop_columns(data, column, inplace = None):
    if inplace == True:
        data.drop(column, axis = 1, inplace = True)
    else:
        data.drop(column, axis = 1)
    return data

# Change Data Type Function 
def changeDtypes(data, cols, types):
    for i in cols:
        data = data.loc[data[i].isin(['1','2','3','4','5','6','7','8','9'])]
        data[i] = data[i].astype(types)
    return data
```

The output above shows there are some prices in the data that are rated Daily and not Annually. <br>
Based on domain knowledge, we could ascertain the minimum threshold of house prices in the areas from which data was collected.<br>
We therefore take out rows that dont meet the least minimum price threshold.

```python
def data_cleaning(data, min_thresh, location, data_name):
    drop_columns(data, 'Unnamed: 0', inplace= True) # .......................... Drop unwanted columns
    data = data.loc[data['Price'] >= min_thresh].reset_index(drop = True) # .... Select prices rated annually
    data['Location'] = data.Location.str.strip() # ............................. Remove extra spaces from the Location feature
    data = changeDtypes(data, ['Bedroom',	'Bathrooms', 'Toilet'], int) # ..... Change datatype
    data = data.loc[data['Location'].isin(location)].reset_index(drop = True) #. Select only desired location
    data.to_csv(f'datasets/production_data/{data_name}_deploy.csv') # ........ Save dataframe
    display(data)
    return data
```

<h3 style = "color:#FF2171"><b><center>Data Preprocessing</b></h3></center><hr style = "width:400px;text-align:center;color:#EDB7ED">

```python
def preprocess(data, encoder_name):
    encoder = LabelEncoder()

    data.dropna(inplace = True)

    # Encode the location area 
    data['Location Area'] = encoder.fit_transform(data['Location Area'])

    # Serialize the encoder 
    joblib.dump(encoder, f'{encoder_name}.pkl')
    return data
```

<h3 style = "color:#FF2171"><b><center>Modelling</b></h3></center><hr style = "width:400px;text-align:center;color:#EDB7ED">

```python
# Run various algorithms to select the best

def modeller(data, show_train_score = False):
    # Split Into Train and Test
    x = data.drop(['Price'], axis = 1)
    y = data['Price']

    global xtrain, xtest, ytrain, ytest
    
    xtrain, xtest, ytrain, ytest = train_test_split(x, y, train_size = 0.90)
    print(f'xtrain: {xtrain.shape}')
    print(f'xtest: {xtest.shape}')
    print('ytrain: {}'.format(ytrain.shape))
    print('ytest: {}'.format(ytest.shape))

    # list of algorithms to be used
    modelling_algorithms = [XGBRegressor(), ElasticNet(), HuberRegressor(), GradientBoostingRegressor(), RandomForestRegressor(), 
                          DecisionTreeRegressor(), LinearRegression(), RANSACRegressor()]
    
    algo_names = ['XGBoost Regressor', 'ElasticNet Regressor', 'Huber Regressor', 'Gradient Booster', 'RandomForest Regressor', 
                  'Decision Regressor', 'Linear Regression', 'Ransac Regressor']
    # keep the training and test score
    test_score = []
    train_score = []

    for algorithm in modelling_algorithms:
        algorithm.fit(xtrain, ytrain)
        model_validate = algorithm.predict(xtrain)
        train_score.append(r2_score(model_validate, ytrain))

        test_perf = algorithm.predict(xtest)
        test_score.append(r2_score(test_perf, ytest))

    # create a table for each algorithms and test_score
    compare_table = pd.DataFrame()
    compare_table['Algorithm'] = pd.Series(algo_names)
    compare_table['Test Accuracy'] = (pd.Series(test_score) * 100).round(2)

    if show_train_score == True:
        compare_table['Training Accuracy'] = (pd.Series(train_score) * 100).round(2)
        compare_table['Overfitting Gap '] = (compare_table['Training Accuracy'] - compare_table['Test Accuracy']) 
        return compare_table.sort_values(by = 'Test Accuracy', ascending = False).reset_index(drop = True)
    else:
        return compare_table.sort_values(by = 'Test Accuracy', ascending = False).reset_index(drop = True)
```

<h5>IKEJA HOUSE PRICE MODELLING <hr>

```python
modeller(ikeja_processed.drop(['Location', 'State'], axis = 1), show_train_score = True)
```

```python
ikeja_model = RandomForestRegressor()
ikeja_model.fit(xtrain, ytrain)
print(f'R2_score of the Model: {r2_score(ikeja_model.predict(xtest), ytest).round(2)}')
print(f'Mean Absolute Error of Model: {mean_absolute_error(ikeja_model.predict(xtest), ytest).round(2)}')
# save model 
pickle.dump(ikeja_model, open('ikeja_model.pkl', "wb"))
```
